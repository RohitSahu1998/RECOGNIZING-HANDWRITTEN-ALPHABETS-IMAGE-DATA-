```python
from PIL import Image
import os
from os import listdir
import numpy as np
import pandas as pd

directory = "/content/sample_data/New folder"
image = []
label = []
for item in os.listdir(directory):
  folder_dir = os.path.join(directory, item)
# get the path/directory

  for i in os.listdir(folder_dir):
       img = Image.open(os.path.join(folder_dir, i))
       image.append(img) # list contains all images
       label.append(i[0]) # contains lables of each image
```

```python
len(image)
```

```python
# Checking the dimension and size of the data
# Before cleaning we have got shape of the images and size is same i.e 28*28(shape) ,784(size)
import matplotlib.pyplot as plt
print(np.array(image[100]).shape)
print(np.array(image[100]).size)
plt.imshow(image[100])
```

```python
# As the Size of the image is large and to resize the dimensionalty we use thumbnail which takes care of the 
# Aspect Ratio as well
# resizing
img_resized=[]
for i in image:
   i.thumbnail((15,15))
   img_resized.append(i)
```

```python
img_resized
```

```python
print(np.array(img_resized[0]).size)  
print(np.array(img_resized[0]).shape)
plt.imshow(img_resized[0])
```

```python
# Now we have to convert the file into numerical form , So for that we have to first convert it into numpy matrix
# As the file is a matrix we have to convert it into a single dimension Array . For that we use flatten method to convert 
# the matrix into a single dimensional Array
l=[]
for i in img_resized:
  img_arr=np.array(i)
  img_rav=img_arr.ravel()
  l.append(list(img_rav)) 
```

```python
mnist_df = pd.DataFrame(l)
mnist_df["label"] = label
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0) 
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train, y_train)
```

```python
from sklearn import metrics

LogisticRegression_score=metrics.accuracy_score(y_test, y_test_pred)
LogisticRegression_score
```
